```python
import pandas as pd
from datetime import datetime
import numpy as np

# Remove extra spaces from column names (both leading, trailing, and internal)
df.columns = df.columns.str.strip().str.replace(r'\s+', '_', regex=True)

# Print the actual column names to verify cleaning
print("Column names after cleaning:")
print(df.columns.tolist())

# List the columns you want to drop
columns_to_drop = ['permalink', 'homepage_url', 'category_list', 'founded_month', 'founded_quarter']

# Drop only the columns that exist in the DataFrame
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns], axis=1)

# Convert 'founded_at' to datetime format
df['founded_at'] = pd.to_datetime(df['founded_at'], errors='coerce')

# Now that 'founded_at' is a datetime object, format it as 'YYYY-MM-DD'
df['founded_at'] = df['founded_at'].dt.strftime('%Y-%m-%d')

# Ensure 'funding_total_usd' is a string before replacing non-numeric values
df['funding_total_usd'] = df['funding_total_usd'].astype(str).str.replace(',', '')

# Replace non-numeric values and convert to float
df['funding_total_usd'] = pd.to_numeric(df['funding_total_usd'].replace(' -   ', np.nan), errors='coerce')

# Remove duplicate rows, if any
df = df.drop_duplicates()

df.head()
```

```python
# Drop rows with NaN values from the entire DataFrame
df = df.dropna()

# Check for missing values in each column after dropping NaN rows
missing_values = df.isnull().sum()

# Print the missing values for each column
print("Missing values in each column after dropping rows with NaN values:")
print(missing_values)

df.shape
```
